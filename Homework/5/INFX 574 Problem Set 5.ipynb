http://scikit-learn.org/stable/modules/model_evaluation.html#defining-your-scoring-strategy-from-metric-functions
the documentation above outlines the creation of a custom loss function for model evaluation

```python
def meanQuadraticLoss(ground_truth, predictions):
    #diff = np.subtract(ground_truth, predictions)
    #result = np.matmul(np.transpose(diff),diff)/ground_truth.size
    return np.average((ground_truth - predictions)**2)
loss = metrics.make_scorer(meanQuadraticLoss, greater_is_better=False)
```

```python
#first model - use RM and ZN features
df_RM_ZN = df[['RM', 'ZN']]
df_RM_ZN.head()
x0 = df_RM_ZN.as_matrix()
y = df[['target']].as_matrix()

lm = linear_model.LinearRegression()
#scores = model_selection.cross_val_score(lm, x0, y, cv=10, scoring=metrics.make_scorer(metrics.mean_squared_error))
scores = model_selection.cross_val_score(lm, x0, y, cv=10, scoring=loss)

scores
```

```python
print("Mean: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
