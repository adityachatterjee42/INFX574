```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
def rot(alpha):
    a = alpha*np.pi/180
    return np.matrix([[np.cos(a), -np.sin(a)], [np.sin(a), np.cos(a)]])

def plotFigure(A):
    plt.figure(figsize=(8,8))
    # set figure size
    plt.grid(b=True)
    # show grid
    min_limit = np.min(A) - np.max(A)*2.0
    max_limit = np.max(A) + np.max(A)*2.0
    plt.axes().set_xlim((min_limit, max_limit))
    plt.axes().set_ylim((min_limit, max_limit))
    # set limits (not the best ones but good enough)
    plt.axes().set_aspect('equal')
    # ensure units are equal along the axes
    plt.plot(A[:,0], A[:,1])
    # plot x, y
    plt.axhline(0, color='black', linestyle = 'dashed')
    plt.axvline(0, color='black',linestyle = 'dashed')
    # draw the axes

A = np.matrix('0 0; 0 1; 0.5 0.5')
plotFigure(A)
```

```python
rot45 = rot(45)
Arot45 = A @ rot45
plotFigure(Arot45)
```

```python
rotm225 = rot(-225)
Arotm225 = A @ rotm225
plotFigure(Arotm225)
```

```python
rotm90 = rot(-90)
Arotm90 = A @ rotm90
plotFigure(Arotm90)
```

```python
B = np.matrix('0 3; -2 -2; 2 2; 2 -2; -2 3')
print(type(B))
plotFigure(B)
```

```python
C = pd.read_csv('crazy_hat.tsv', sep='\t')
Hat = C.as_matrix()
print(Hat)
```

```python
Hat=Hat[:,:2]
plotFigure(Hat)
```

```python
Hat = C
Hat.columns=['x','y','group']
print(Hat)
print(Hat['group'].unique())
```

```python
groups=[]
for group in Hat['group'].unique():
    groups.append(Hat[Hat['group']==group])
groups_mx = list(map(lambda x: x.as_matrix()[:,:2], groups))
#mx = Hat.as_matrix()[:,:2]
mx = C.as_matrix()[:,:2]

def plotFigureList(mx, groups_mx):
    plt.figure(figsize=(8,8))
    # set figure size
    plt.grid(b=True)
    # show grid
    min_limit = np.min(mx) - np.max(mx)*2.0
    max_limit = np.max(mx) + np.max(mx)*2.0
    plt.axes().set_xlim((min_limit, max_limit))
    plt.axes().set_ylim((min_limit, max_limit))
    # set limits (not the best ones but good enough)
    plt.axes().set_aspect('equal')
    # ensure units are equal along the axes
    for x in groups_mx:
        plt.plot(x[:,0], x[:,1])
    # plot x, y
    plt.axhline(0, color='black', linestyle = 'dashed')
    plt.axvline(0, color='black',linestyle = 'dashed')
    # draw the axes
    
plotFigureList(mx, groups_mx)
```

```python
r=40
mx=mx.astype(int)
mxrotr = mx @ rot(r)
groups_mxrotr = list(map(lambda x: x.astype(int) @rot(r) , groups_mx))
plotFigureList(mxrotr, groups_mxrotr)
```
